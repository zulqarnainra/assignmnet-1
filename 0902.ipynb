# Question 1

```python
import pandas as pd
import numpy as np
import matplotlib as plt
from matplotlib import pyplot
df=pd.read_csv('jsrt_metadata.csv')
```

```python
count=df['diagnosis'].value_counts()
lbl=df['diagnosis'].value_counts().index.tolist()
count
```

```python
#Method 1
pyplot.barh(lbl,count,color='red')
```

```python
#Method 2
plt.pyplot.title("Histogram")
plt.pyplot.xlabel("Diseases")
plt.pyplot.ylabel("Frequency")
count.plot(kind='bar', figsize=(14,6))
```
